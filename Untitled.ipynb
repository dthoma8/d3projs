```python
import pandas as pd
```

```python
av_data = pd.read_csv("./data/Availability.csv")
av_data.head()
```

```python
def data2CsvString(data):
    
    string = ",".join(data.columns.tolist())+"\n"
    for i, row in data.iterrows():
        
        obj = row.astype(str).tolist()
        string=string+",".join(obj[:len(obj)-1])+',"%s"\n'%obj[-1]
        
    return string

def writeCsv(csvstring, fname):
    
    with open("./data/"+fname+".csv", "w") as f:
        f.write(csvstring)
        f.close()
```

```python
av_data_string = data2CsvString(av_data)
av_data_string
```

```python
writeCsv(av_data_string,)
```
